### Business Case studies
#### Flight overbooking problem

```python
import math
PENALTY=10000

def calculate_expected_probablity(ticket_sold):
    total_penalty=0
    for i in range(1, ticket_sold - 100+1):
        prob= math.comb(ticket_sold, 100 + i)* (0.9)**(100+i) * (0.1)**(ticket_sold-100-i)
        penalty=prob*PENALTY*i
        # print(f"probability of {100+i} customer showing: {prob}")
        total_penalty+=penalty
    return total_penalty
```

```python
calculate_expected_probablity(110)
```

```python
for i in range(100,130):
    sales=5000*i
    penalty=calculate_expected_probablity(i)
    profit=sales-penalty
    print("Overbooking Seats {}, Profit made: {} ".format(i, round(profit)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create data points for x-axis
x = np.linspace(55, 85, 1000)  # Range from μ-3σ to μ+3σ

# Calculate PDF and CDF values
pdf_values = norm.pdf(x, loc=70, scale=5)
cdf_values = norm.cdf(x, loc=70, scale=5)

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
fig.suptitle('Normal Distribution (μ=70, σ=5)', fontsize=14)

# Plot PDF
ax1.plot(x, pdf_values, 'b-', lw=2, label='PDF')
ax1.fill_between(x, pdf_values, alpha=0.3)
ax1.axvline(x=70, color='r', linestyle='--', label='Mean (μ=70)')
ax1.grid(True, alpha=0.3)
ax1.set_title('Probability Density Function (PDF)')
ax1.set_ylabel('Density')
ax1.legend()

# Add annotations for standard deviations on PDF
for i in [-2, -1, 1, 2]:
    ax1.axvline(x=70 + i*5, color='g', linestyle=':', alpha=0.5)
    ax1.text(70 + i*5, max(pdf_values)/2, f'{i}σ', rotation=90)

# Plot CDF
ax2.plot(x, cdf_values, 'b-', lw=2, label='CDF')
ax2.axvline(x=70, color='r', linestyle='--', label='Mean (μ=70)')
ax2.axhline(y=0.5, color='g', linestyle=':', label='Median (0.5)')
ax2.grid(True, alpha=0.3)
ax2.set_title('Cumulative Distribution Function (CDF)')
ax2.set_xlabel('Value')
ax2.set_ylabel('Probability')
ax2.legend()

# Add annotations for important probabilities on CDF
ax2.text(71, 0.5, '50%', verticalalignment='bottom')
ax2.text(75, 0.84, '84%', verticalalignment='bottom')
ax2.text(65, 0.16, '16%', verticalalignment='bottom')

plt.tight_layout()
plt.show()
```
