```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from scipy import stats
```

```python
uniform_data=stats.uniform.rvs(loc=1, scale=100, size=10000)
```

```python
sns.histplot(uniform_data)
plt.show()
```

```python
sns.distplot(uniform_data)
plt.show()
```
