### Bar Plot

```python
sales_nano=[24,19,15,11,8]
sales_swift=[15,17,23,11,13]
```

```python
sales_nano=np.array(sales_nano)
sales_swift=np.array(sales_swift)
```

```python
x=np.arange(2017,2022,1)
```

```python
x
```

```python
plt.bar(x,sales_swift+1,label="swift", alpha=0.5,width=0.3)
plt.bar(x+0.3,sales_nano+1,label="nano",alpha=0.6,width=0.3)
plt.xlabel("Years")
plt.ylabel("Sales")
plt.legend()
plt.show
```

## Pie Chart
