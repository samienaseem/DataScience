### Bootstraping in Python

```python
np.random.choice(df_sample['travel_time'],size=10)
```

```python
r=10000
```

```python
sample=df_sample['travel_time'].values
```

```python
size=sample.shape[0]
```

```python
bs_samples=[]
```

```python
for i in range(r):
    bs_sample=np.random.choice(sample, size=size)
    bs_samples.append(bs_sample.mean())
```

```python
bs_samples
```

```python
bs_samples=np.array(bs_samples)
```

```python
sns.histplot(bs_samples,bins=25, kde=True)
```

### Exponential Distribution 

```python
ex_sample=np.random.exponential(scale=10,size=1000)
```

```python
plt.hist(ex_sample,bins=25
```

```python
r=1000
size=len(ex_sample)

ex_samples=np.empty(r)

for i in range(r):
    bs_sample=np.random.choice(ex_sample,size=size)
    ex_samples[i]=bs_sample.mean()

plt.figure()
sns.histplot(ex_samples,bins=25,kde=True)
plt.title("Sampling distribution of means from exponential distribution samples")
plt.show()
```
