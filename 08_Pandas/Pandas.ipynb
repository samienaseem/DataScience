### Feature Engineering

```python
data.dtypes
```

```python
data['START_DATE']
```

```python
#to convert object to datetime 
data['START_DATE']=pd.to_datetime(data['START_DATE'])
data['END_DATE']=pd.to_datetime(data['END_DATE'])
```

```python
data["START_DATE"].head()
```

```python
data.dtypes
```

```python
date=data["START_DATE"][0]
```

```python
### things we can extract from timnstamp
```

```python
print(date.year)
print(date.month)
print(date.month_name())
print(date.day_name())
print(date.dayofweek)
print(date.hour)
```

```python
hour=[]
```

```python
year=[]
month_name=[]
month=[]
day=[]
day_name=[]
day_of_week=[]
hour=[]
```

```python
for i in data["START_DATE"]:
    month_name.append(i.month_name())
    day.append(i.day)
    day_name.append(i.day_name())
    day_of_week.append(i.dayofweek)
    hour.append(i.hour)
    
```

```python
for i in data["START_DATE"]:
    #hour.append(i.hour)
    year.append(i.year)
    month.append(i.month)
    
```

```python
data["YEAR"]=year
data["MONTH"]=month
data['MONTH_NAME']=month_name
data["DAY"]=day
data["DAY_NAME"]=day_name
data["DAY_OF_WEEK"]=day_of_week
data["HOUR"]=hour
```

```python
data.head()
```
